```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
def rollDice():
    roll = np.random.randint(1, 101)
    return 50 < roll < 100
```

```python
def doubler_bettor(funds, initial_wager, wager_count):
    value_results = [funds]
    
    value = funds
    wager = initial_wager
    
    previous_win = True
    
    for i in range(wager_count):
        if rollDice():
            value += wager
            previous_win = True
            wager = initial_wager
        else:
            value -= wager
            previous_win = False
            wager = wager*2
        
        value_results.append(value)
    
    return value_results
```

```python
df = pd.DataFrame()
for col in range(10):
    df[col] = doubler_bettor(10000, 100, 500)
df.plot()
```

```python
def doubler_bettor_broke(funds, initial_wager, wager_count):
    value_results = [funds]
    
    value = funds
    wager = initial_wager
    
    previous_win = True
    
    for i in range(wager_count):
        if value <= 0:
            return value_results
        if rollDice():
            value += wager
            previous_win = True
            wager = initial_wager
        else:
            value -= wager
            previous_win = False
            wager = wager*2
        
        value_results.append(value)
    
    return value_results
```

```python
df = pd.DataFrame(index = range(501))
for col in range(10):
    array = doubler_bettor_broke(10000, 100, 500)
    df.loc[range(len(array)),col] = array
df.plot()
plt.axhline(0, ls = 'dashed')
```
