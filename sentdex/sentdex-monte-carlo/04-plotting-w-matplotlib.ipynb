```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
def simple_bettor_n(funds, initial_wager, wager_count, n):
    rolls = np.random.randint(1, 101, size = [wager_count, n])
    winnings = initial_wager*(rolls>50)*(rolls<100)
    losses = -(initial_wager*(rolls<=50) + initial_wager*(rolls == 100))
    return funds + np.cumsum(winnings + losses, axis = 0)
```

```python
df_results = pd.DataFrame(simple_bettor_n(10000, 100, 100, 10))
df_results.plot()
```

```python
df_results = pd.DataFrame(simple_bettor_n(10000, 100, 1000, 10))
df_results.plot()
```
