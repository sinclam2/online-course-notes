```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
def rollDice():
    roll = np.random.randint(1, 101)
    return 50 < roll < 100
```

```python
def doubler_bettor_broke(funds, initial_wager, wager_count):
    value_results = [funds]
    
    value = funds
    wager = initial_wager
    
    previous_win = True
    
    for i in range(wager_count):
        if value <= 0:
            return value_results
        if rollDice():
            value += wager
            previous_win = True
            wager = initial_wager
        else:
            value -= wager
            previous_win = False
            wager = wager*2
            #adding line to fix going into debt
            if (value - wager) < 0:
                wager = value

        value_results.append(value)
    
    return value_results
```

- Now we'll only be wagering what we have left

```python
df = pd.DataFrame(index = range(1001))
for col in range(10):
    array = doubler_bettor_broke(10000, 100, 1000)
    df.loc[range(len(array)),col] = array
```

```python
df.plot()
```
